```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Mall_Customers.csv")
```

```python
df.head()
```

```python
print(df.head())
```

```python
# Sactter Plot
sns.scatterplot(x="Age", y="Spending Score (1-100)", hue="Genre", data=df, alpha=0.7)
plt.title("Age vs Spending Score")
plt.show()

```

```python
# Bar Plot
sns.barplot(x="Genre", y="Spending Score (1-100)", data=df, estimator=sum)
plt.title("Total Spending Score by Genre")
plt.show()

```

```python
# Box Plot
sns.boxplot(x="Genre", y="Annual Income (k$)", data=df)
plt.title("Box Plot: Annual Income by Genre")
plt.show()

```

```python
# Pie Charts
plt.pie(df["Genre"].value_counts(), labels=df["Genre"].unique(), autopct="%1.1f%%", colors=["red", "purple"])
plt.title("Pie Chart: Gender Distribution")
plt.show()

```

```python
# Line chart
plt.plot(df["CustomerID"], df["Spending Score (1-100)"], "o-", color="purple")
plt.title("Line Chart: Spending Score Over Customer IDs")
plt.xlabel("Age")
plt.ylabel("Spending Score (1-100)")
plt.show()

```
